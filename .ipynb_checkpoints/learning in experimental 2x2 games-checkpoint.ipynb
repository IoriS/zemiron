```julia
function selfbattle(model,games)
    lastresult = zeros(12,2)
    for number in 1:12
        a = games[2*number-1,1]
        b = games[2*number-1,2]
        c = games[2*number,1]
        d = games[2*number,2]
        result = zeros(2)
        payoff_matrix = [a b;c d]
        for repeat in 1:500
            gameresult =model(payoff_matrix)
            result[1] += gameresult[1]
            result[2] += gameresult[2]
        end
        lastresult[number,1] += (result[1])/500
        lastresult[number,2] += (result[2])/500
    end
    return lastresult
end
```

```julia
game1=[(10,8) (0,18);(9,9) (10,8)]
game2=[(9,4) (0,13);(6,7) (8,5)]
game3=[(8,6) (0,14);(7,7) (10,4)]
game4=[(7,4) (0,11);(5,6) (9,2)]
game5=[(7,2) (0,9);(4,5) (8,1)]
game6=[(7,1) (1,7);(3,5) (8,0)]
game7=[(10,12) (4,22);(9,9) (14,8)]
game8=[(9,7) (3,16);(6,7) (11,5)]
game9=[(8,9) (3,17);(7,7) (13,4)]
game10=[(7,6) (2,13);(5,6) (11,2)]
game11=[(7,4) (2,11);(4,5) (10,1)]
game12=[(7,3) (3,9);(3,5) (10,0)]
games=[game1;game2;game3;game4;game5;game6;game7;game8;game9;game10;game11;game12]
```

シミュレーションの結果と観察結果との距離の二乗の平均値を出す。

```julia
function quadraticdistance(model,games,data)
    lastresult = 0
    for number in 1:12
        a = games[2*number-1,1]
        b = games[2*number-1,2]
        c = games[2*number,1]
        d = games[2*number,2]
        gameresult = 0
        payoff_matrix = [a b;c d]
        for repeat in 1:500
            result=model(payoff_matrix)
            gameresult+=(result[1]-data[number,1])^2 +(result[2]-data[number,2])^2
        end
        lastresult += (gameresult)/500
    end
    return lastresult/12
end
```

```julia
quadraticdistance(impulsematching,games,data)
```

```julia
quadraticdistance(actionsampling,games,data)
```

```julia
quadraticdistance(reinforcement,games,data)
```

```julia
quadraticdistance(stEWA,games,data)
```

impulse-matching learningで行われている利得変形を他の学習モデルでも試す。

```julia
include("transformer.jl")
```

```julia
newgame = Array{Any}(12)
for w in 1:12
    a = games[2*w-1,1]
    b = games[2*w-1,2]
    c = games[2*w,1]
    d = games[2*w,2]
    payoff_matrix = [a b;c d]
    newgame[w] =transformer(payoff_matrix)
end
transformed_games=[newgame[1];newgame[2];newgame[3];newgame[4];newgame[5];newgame[6];newgame[7];newgame[8];newgame[9];newgame[10];newgame[11];newgame[12]]
```

loss-aversionの程度を時間に依存させてシミュレーションを行う。1が徐々に損失回避的になるもの、2が最初が損失回避的で徐々に元の利得を重視するもの、3が最初は損失回避的で中盤では程度が下がり、終盤にかけて再び損失回避的になるものである。

```julia
include("LAchange_impulsematching_1.jl")
include("LAchange_impulsematching_2.jl")
include("LAchange_impulsematching_3.jl")
```

```julia
function paramchange_impulse_quad(model,alpha,beta,games,data)
    lastresult = 0
    for number in 1:12
        a = games[2*number-1,1]
        b = games[2*number-1,2]
        c = games[2*number,1]
        d = games[2*number,2]
        gameresult = 0
        payoff_matrix = [a b;c d]
        for repeat in 1:500
            result=model(payoff_matrix,alpha,beta)
            gameresult+=(result[1]-data[number,1])^2 +(result[2]-data[number,2])^2
        end
        lastresult += (gameresult)/500
    end
    lastresult = (lastresult)/12
end
```

```julia
function param_est(n,games,data)
    lastresult = 0
    for number in 1:12
        a = games[2*number-1,1]
        b = games[2*number-1,2]
        c = games[2*number,1]
        d = games[2*number,2]
        gameresult = 0
        payoff_matrix = [a b;c d]
        for repeat in 1:500
            result=impulsematching(payoff_matrix,n)
            gameresult+=(result[1]-data[number,1])^2 +(result[2]-data[number,2])^2
        end
        lastresult += (gameresult)/500
    end
    lastresult = (lastresult)/12
end
```

```julia
x_3 = collect(linspace(1,10,200))
y_3 = zeros(length(x_3))
for i in 1:length(x_3)
    y_3[i] = param_est(x_3[i],games,data)
end
plot(x,y,"b-",linewidth = 2)
```
